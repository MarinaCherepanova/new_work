```python
df = pd.read_csv('iris.data', names = ['1','2','3','4','Category'])
df.hist(bins = 150, figsize = (15,9))
```

```python
unique_arr = np.unique(category_arr)
unique_arr
```

```python
out = []
for category in unique_arr:
    for a in special_arr[special_arr == category]:
        group_id = np.argwhere(unique_arr == a).tolist()[0][0]
        out.append(group_id)
print(out)
```

```python
out_arr = np.array(out)
out_arr = out_arr.reshape((150,1))
```

```python
main_arr = np.concatenate((data, out_arr), axis = 1)
main_arr
```

```python
category_index = main_arr[:,5].astype('int')
value = main_arr[:,0].astype('float')
result = []

for category in np.unique(category_index):
    result.append([category,
                   value[category_index == category].min(), 
                   value[category_index == category].max(), 
                   np.median(value[category_index == category]), 
                   value[category_index == category].mean(), 
                   np.percentile(value[category_index == category], 25), 
                   np.percentile(value[category_index == category], 75),
                   ])
    
df.hist(by = "Category", figsize = (12,6))
```
